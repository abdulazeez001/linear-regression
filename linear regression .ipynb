```python
ex1data  = np.loadtxt(r'C:\Users\UNION\Downloads\ex1data1.txt', delimiter=',')
```

```python
x = np.c_[ex1data[:,0],np.ones(ex1data.shape[0])]
```

```python
y = np.c_[ex1data[:,1]]
```

```python
import matplotlib.pyplot as plt
plt.scatter(x[:,0],y)
```

```python
def train(x,y):
    m = y.size
    iteration =5000
    theta = [[0],[0]]
    cost_function = np.zeros(iteration)
    
    for i in range(iteration):
        h = x @ theta
        theta = update_weight(h,theta,m,x)
        cost_function[i] = cost(x,theta,m)
        print("the cost_function[",i,"] is " ,cost_function[i])
    return theta,cost_function

def update_weight(h,theta,m,x):
    alpha = 0.01
    theta = theta - alpha * (1/m * (x.T @ (h - y )))
    return theta

def cost(x,theta,m):
    h = x @ theta
    cost_value = 0
    cost_value = (1/2*m) * np.sum(np.square(h-y))
    return cost_value
    
        
```

```python
 theta,cost_value=train(x,y)
```

```python
plt.plot(cost_value)
plt.ylabel('cost J')
plt.xlabel('iteration')
```
